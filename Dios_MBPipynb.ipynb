numbers of rating dan battery mah tetap disimpan outliernya karena data tersebut memang nyata dan masuk akal

```python
num_cols = ['Number of Ratings', 'Rating', 'RAM (GB)', 'ROM (GB)', 'Battery (mAh)', 'Back Cam MP', 'Jumlah Back Cam', 'Front Cam MP', 'Price in INR']

# identifikasi dan menghandle
def handle_outliers(df, col, method='remove'):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    if method == 'remove':
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    elif method == 'transform':
        df[col] = np.clip(df[col], lower_bound, upper_bound)
    return df

# membagi kategori outliers
cols_to_remove_outliers = ['Rating', 'RAM (GB)', 'ROM (GB)', 'Back Cam MP', 'Jumlah Back Cam', 'Front Cam MP']

cols_to_transform_outliers = ['Price in INR']

cols_to_keep_outliers = ['Number of Ratings', 'Battery (mAh)']


for col in cols_to_remove_outliers:
    df = handle_outliers(df, col, method='remove')

for col in cols_to_transform_outliers:
    df = handle_outliers(df, col, method='transform')


plt.figure(figsize=(12,8))
sns.boxplot(data=df[num_cols])
plt.title('Boxplots after outlier handling')
plt.show()

df[num_cols].hist(bins=30, figsize=(12,8))
plt.title('Histograms after outlier handling')
plt.show()

print(f'Baris data setelah penanganan outlier: {len(df)}')
```
